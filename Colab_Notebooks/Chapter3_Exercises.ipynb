```python
#Chapter 3: Challenge
def get_grade(score):
    if 97 <= score <= 100:
        return 1.0
    elif 93 <= score < 97:
        return 1.25
    elif 89 <= score < 93:
        return 1.5
    elif 85 <= score < 89:
        return 1.75
    elif 81 <= score < 85:
        return 2.0
    elif 77 <= score < 81:
        return 2.25
    elif 74 <= score < 77:
        return 2.5
    elif 0 <= score < 74:
        return 3.0
    else:
        return None

def main():
    while True:
        try:
            score = float(input("Enter a score between 0 and 100: "))
            if 0 <= score <= 100:
                grade = get_grade(score)
                if grade is not None:
                    print("Grade:", grade)
                else:
                    print("Invalid score entered. Please try again.")
            else:
                print("Score must be between 0 and 100.")
        except ValueError:
            print("Invalid input. Please enter a valid number.")

if __name__ == "__main__":
    main()
```
