```python
#Chapter4: Exercise 1
def compute_pay(hours, rate):
    if hours <= 40:
        pay = hours * rate
    else:
        regular_pay = 40 * rate
        overtime_pay = (hours - 40) * (rate * 1.5)
        pay = regular_pay + overtime_pay
    return pay

def main():
    try:
        hours = float(input("Enter Hours: "))
        rate = float(input("Enter Rate: "))
        if hours >= 0 and rate >= 0:
            pay = compute_pay(hours, rate)
            print("Pay:", pay)
        else:
            print("Hours and rate must be non-negative numbers.")
    except ValueError:
        print("Invalid input. Please enter valid numbers.")

if __name__ == "__main__":
    main()
```

```python
#Chapter 4: Challenge
def computegrade(score):
    if 97 <= score <= 100:
        return "1.0"
    elif 93 <= score < 97:
        return "1.25"
    elif 89 <= score < 93:
        return "1.5"
    elif 85 <= score < 89:
        return "1.75"
    elif 81 <= score < 85:
        return "2.0"
    elif 77 <= score < 81:
        return "2.25"
    elif 74 <= score < 77:
        return "2.5"
    elif 0 <= score < 74:
        return "3.0"
    else:
        return "Invalid score"

def main():
    while True:
        try:
            score = float(input("Enter a score between 0 and 100: "))
            if 0 <= score <= 100:
                grade = computegrade(score)
                print("Grade:", grade)
            else:
                print("Score must be between 0 and 100.")
        except ValueError:
            print("Invalid input. Please enter a valid number.")

if __name__ == "__main__":
    main()
```
