```python
#Chapter 8 - Lists
#Exercise 1
numbers = [] #Create an empty list

#Add numbers using append
numbers.append(4)
numbers.append(6)
numbers.append(9)

print(numbers)  #Output the list
```

```python
#Exercise 2
numbers = list(range(11)) #Create a list containing numbers from 0 to 10 using range

print(numbers)  #Output the list
```

```python
#Exercise 3
grades = [1.0, 1.25, 1.5, 1.75, 2.0, 2.25, 2.5, 3.0] #Create a list containing valid grade values

for index, grade in enumerate(grades): #Iterate over the list using enumerate
    print(index, grade)
```

```python
#Exercise 4
grades = [1.0, 1.25, 1.5, 1.75, 2.0, 2.25, 2.5, 3.0] #Create a list containing valid grade values

grades_above_2 = grades[grades.index(2.25):] #Use index slicing to print out grades higher than 2.0
print(grades_above_2)
```

```python
#Exercise 5
grades = [1.0, 1.25, 1.5, 1.75, 2.0, 2.25, 2.5, 3.0] #Define the list of grades

index_2 = grades.index(2.0) #Find the index of the grade 2.0
better_than_2 = grades[index_2 + 1:]#Slice the list from the index next to 2.0 to the end

print(better_than_2) #Print the list of grades better than 2.0
```

```python
#Exercise 6
grades_list = [1.0, 1.25, 1.5, 1.75, 2.0, 2.25, 2.5, 3.0]  #Define the list of grades

print(grades_list[grades_list.index(2.0):grades_list.index(3.0)])  #Using index slicing to print out the grades 2.0, 2.25, and 2.5
```
