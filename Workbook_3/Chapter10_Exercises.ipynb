```python
#Exercise 2
#Define the tuple containing information about the Philippines
ph_info = ('Philippines', 'Manila', 103.3)  #Country, Capital, Population in Millions

#Unpack the tuple into meaningful variables
country, capital, population_millions = ph_info

#Display the unpacked information
print("Country:", country)
print("Capital:", capital)
print("Population (in millions, as of 2016):", population_millions)
```
