```python
#Chapter 5 - Iterations
#Chapter 5: Exercise 1
a = 2
b = 2.0
c = a

print(a is c)  # True, because a and c both refer to the integer object 2.
print(a == c)  # True, because the values of a and c are equal.

print(a is b)  # False, because a and b are different types (int and float) and hence cannot point to the same object.
print(a == b)  # True, because although a and b are of different types, their values are equal.
```

```python
#Chapter 5: Challenge 1
total = 0
count = 0

while True:
    try:
        user_input = input("Enter a number: ")
        if user_input.lower() == "done":
            break
        number = float(user_input)
        total += number
        count += 1
    except ValueError:
        print("Invalid input")9
        continue

if count == 0:
    print("No numbers entered.")
else:
    average = total / count
    print(total, count, average)
```

```python
#Chapter 5: Challenge 2
numbers = []

while True:
    try:
        user_input = input("Enter a number: ")
        if user_input.lower() == "done":
            break
        number = float(user_input)
        numbers.append(number)
    except ValueError:
        print("Invalid input. Please enter a valid number or 'done' to finish.")

if len(numbers) == 0:
    print("No numbers entered.")
else:
    max_number = max(numbers)
    min_number = min(numbers)
    print("Maximum:", max_number)
    print("Minimum:", min_number)
```
