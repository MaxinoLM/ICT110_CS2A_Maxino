```python
# string library

#lower case
text = 'Enero, Febrero, Marzo'
spanish_months = text.lower()
print(spanish_months)
##my trial
numbers = 'One, Two, Three'
num = numbers.lower()
print (num)

#upper case
greeting = 'Hello Bob'
greeting.upper()
##mytrial
name = 'vijay'
name.upper()

#stripping whitespaces
greeting = '   Hello Bob  '
greeting.lstrip() #left
greeting.rstrip() #right
greeting.strip() #both sides

#splitting the words
quotation = 'One ring to rule them all, one ring to find them.'
quotation.split()
```
