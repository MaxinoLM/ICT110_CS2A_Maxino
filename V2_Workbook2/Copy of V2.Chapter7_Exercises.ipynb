```python
#Chapter 7 - Files
#Chapter 7: Exercise 1
def print_file_in_uppercase(filename):
    try:
        with open(filename, 'r') as file:
            for line in file:
                print(line.upper(), end='')
    except FileNotFoundError:
        print("File not found.")
```

```python
#Chapter 7: Exercise 2
def count_lines(filename):
    try:
        with open(filename, 'r') as file:
            line_count = 0
            for line in file:
                line_count += 1
            print("Number of lines in the file:", line_count)
    except FileNotFoundError:
        print("File not found.")
```

```python
#Chapter 7: Challenge

# Prompting user for the file name
filename = input("Enter the file name: ")

try:
    # Opening the file
    with open(filename, 'r') as file:
        total_confidence = 0
        count = 0

        # Reading each line in the file
        for line in file:
            # Checking if the line starts with "X-DSPAM-Confidence:"
            if line.startswith("X-DSPAM-Confidence:"):
                # Extracting the floating-point number
                confidence_index = line.find(':') + 1
                confidence = float(line[confidence_index:])

                # Updating total confidence and count
                total_confidence += confidence
                count += 1

        if count > 0:
            # Calculating average spam confidence
            average_confidence = total_confidence / count
            print("Average spam confidence:", average_confidence)
        else:
            print("No lines starting with 'X-DSPAM-Confidence:' found in the file.")

except FileNotFoundError:
    print("File not found. Please enter a valid file name.")
except Exception as e:
    print("An error occurred:", e)
```
